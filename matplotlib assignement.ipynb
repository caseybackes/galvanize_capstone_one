```python
a = np.random.randint(1000,size =50)
b = np.random.randint(1000,size =50)
ab=zip(a,b)
color=list(['r','b'])
mask = [(x[0]+x[1])%2 for x in ab]
mask
color_list = [color[c] for c in mask]
color_list
```

```python
fig1 = plt.figure(figsize=(8,4), dpi=120)
ax1 = fig1.add_subplot(2,2,1)
ax1.scatter(a,b, c=color_list)
#ax.color_bar()

x = np.linspace(0,5,100)
y1 = 3*x + 0.5
y2 = 5 * np.sqrt(x)

#fig2 = plt.figure(figsize = (8,4), dpi=120)
ax2 = fig1.add_subplot(2,2,2)
ax2.scatter(x,y1,color='r', label ='y1')
ax2.plot(x,y2, c='b', label ='y2')
ax2.legend();
ax2.set_yscale('log')


barheights = [3,5,1]
barlabels = ['grapes', 'oranges', 'hockey pucks']
ax3 = fig1.add_subplot(2,2,3)
ax3.bar(np.arange(len(barheights)), barheights)
x_pos = np.arange(len(barheights))
ax3.set_xticks(x_pos)
ax3.set_xticklabels(barlabels, rotation=45)

plt.savefig('matplotlib_plotting_breakout.jpg')
```
